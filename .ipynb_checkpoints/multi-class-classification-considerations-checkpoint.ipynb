<h1>Multi-Class Activation Functions</h1>
<p>
    Notice all these activation functions make sense for a single output, either a continuous label or trying to predict a binary classification (either a 0 or 1).
    <br>But what should we do if we have a multi-class situation?
    <br>There are 2 main types of multi-class situations
    <li>Non-Exclusive Classes</li>
    <li>> A data point can have multiple classes/categories assigned to it</li>
    <li>Mutually Exclusive Classes</li>
    <li>Only one class per data point.</li>
</p>
<p>
    Non-Exclusive Classes
    <br>A data point can have multiple classes/categories assigned to it.
    <br>Photos can have multiple tags (e. g. beach, family, vacation, etc...)
</p>
<p>
    Mutually Exclusive Classes
    <br>A data point can only have one class/category assigned to it
    <br>Photos can be categorized as being in gryscale (black and white) or full color photos. A photo can not be both at the same time.
</p>
<p>
    Organizing Multiple Classes
    <br>The easiest way to organize multiple classes is to simply have 1 output node per class
    <br>Previously we thought of the last output layer as a single node.
    <br>This single node could output a continuous regression value or binary classification (0 or 1).
    <br>Organizing Multiple Classes means we will need to organize categories for this output layer.
    <br>We can't just have categories like "red", "blue", "green", etc...
    <br>Instead we use <strong>one-hot encoding</strong>
    <br>Let's take a look at what this looks like for mutually exclusive classes.
</p>

<p>
    Now that we have our data correclty organized, we just need to choose the correct classification activation function that the last output layer should have
    <br>Non-exclusive
    <li>Sigmoid function: Each neuron will output a value between 0 and 1, indicating the probability of having that class assigned to it</li>
    <li>> Keep in mind this allows each neuron to output independents of the other classes, allowing for a single data point def into the function to have multiple classes assigned to it.</li>
    <br>Mutually Exclusive Classes
    <li>But what do we do when each data point can onlu have a single class assigned to it?</li>
    <li>We can use the <strong>softmax function</strong> for this!</li>
    <li>Softmax function calculate the probabilities distribution of the event over <strong>K</strong> different events.</li>
    <li>This function will calculate the probabilities of each target class over all possible target classes</li>
    <li>The range will be 0 to 1, and <strong>the sum of all the probabilities will be equal to one</strong>.</li>
    <li>The model return the probabilities of each class and the target class chosen will have the highest probability</li>
    <li>The main thing to keep in mind is that if you use softmax for multi-class problems you get this sort of output:</li>
    <li>> [Red, Green, Blue]</li>
    <li>> [0.1, 0.6, 0.3]</li>
    <li>The probabilities for each class all sum up to 1. We choose the highest probability as our assignment.</li>
